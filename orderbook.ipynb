```python
import ccxt
import pandas as pd
import time

pd.set_option('display.max_columns', None)

# ccxtのインスタンス化
exchange = ccxt.bybit()
symbol = 'BTC/USDT'

# 空のデータフレームを作成
order_book_df = pd.DataFrame(columns=['timestamp', 'bids', 'asks'])

# 繰り返し板情報を取得
for _ in range(5):
    # 板情報を取得
    order_book = exchange.fetch_order_book(symbol)
    print(order_book)
    
    # timestamp, bids, asksを取得
    timestamp = order_book['timestamp']
    bids = order_book['bids']
    asks = order_book['asks']
    
    # 取得したデータをデータフレームに追加
    temp_df = pd.DataFrame({'timestamp': [timestamp], 'bids': [bids], 'asks': [asks]})
    order_book_df = pd.concat([order_book_df, temp_df], ignore_index=True)
    
    # 時間をおいて次の板情報を取得
    time.sleep(1)

display(order_book_df)
```
